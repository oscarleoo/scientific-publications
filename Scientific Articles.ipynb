```python
import imageio

import numpy as np
import pandas as pd
import geopandas as gpd

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

from PIL import Image
from matplotlib.patches import Patch
from matplotlib.ticker import FuncFormatter
from matplotlib.colors import LinearSegmentedColormap, ListedColormap
```

## Helper Functions

```python
def set_seaborn_style(font_family, background_color, grid_color, text_color):
    sns.set_style({
        "axes.facecolor": background_color,
        "figure.facecolor": background_color,
        "axes.edgecolor": grid_color,
        "axes.grid": True,
        "axes.axisbelow": True,
        "grid.color": grid_color,
        "text.color": text_color,
        "font.family": font_family,
        
        "xtick.color": text_color,
        "ytick.color": text_color,
        "xtick.bottom": False,
        "xtick.top": False,
        "ytick.left": False,
        "ytick.right": False,

        "axes.spines.left": False,
        "axes.spines.bottom": True,
        "axes.spines.right": False,
        "axes.spines.top": False,
    }
)
    
def add_padding_to_chart(chart, left, top, right, bottom, background):
    size = chart.size
    image = Image.new("RGB", (size[0] + left + right, size[1] + top + bottom), background)
    image.paste(chart, (left, top))
    return image


def create_bar_plot_image(df, year, color, title):
    fig = plt.figure(figsize=(16, 18))
    ax = sns.barplot(y=df.country_name[:20], x=df[year][:20], color=color, orient="h")

    ax.set(xlabel=None, ylabel=None)
    ax.tick_params("y", labelsize=20, pad=8)
    ax.tick_params("x", labelsize=20, pad=16)
    ax.get_xaxis().set_major_formatter(FuncFormatter(lambda x, p: format(int(x), ',')))

    plt.title(title, fontsize=32, y=1.06, x=0.48)
    plt.tight_layout()

    fig.canvas.draw()
    data = np.frombuffer(fig.canvas.tostring_rgb(), dtype=np.uint8)
    data = data.reshape((fig.canvas.get_width_height()[::-1]) + (3,))
    plt.close() 

    chart_image = Image.fromarray(data)
    chart_image = add_padding_to_chart(chart_image, 30, 50, 100, 30, background_color)
    return chart_image
```

## About the Data

The data measure the number of scientific articles published in a set of journals. If an article have multiple authors that comes from different countries each country get a fraction. We have data from 1996 to 2020 for every country in the world.

```python
df = pd.read_csv("./data/scientific.csv", index_col=None)
population = pd.read_csv("./data/population.csv", index_col=None)

df = df.sort_values("country_name")
population = population.sort_values("country_name")
```

```python
population.head()
```

We convert the population data to number of scientific articles per 100,000.

```python
years = list(population.columns[1:])
population.iloc[:, 1:] = df[years] * (100000 / population[years])

df = df.sort_values("2020", ascending=False)
population = population.sort_values("2020", ascending=False)
```

```python
df = df.sort_values("2020", ascending=False)
population = population.sort_values("2020", ascending=False)
```

## Setting up Seaborn

```python
font_family = "PT Mono"
background_color = "#FAE8E0"
text_color = "#040F0F"
grid_color = "#DDC9A7"

set_seaborn_style(font_family, background_color, grid_color, text_color)
```

```python
color_palette = sns.color_palette([
    "#EF7C8E",
    "#99D6C2",
    "#85A5EA",
])
```

```python
color_palette
```

## Most Scientific Publications

```python
create_bar_plot_image(
    df=df,
    year="2020",
    color=color_palette[0],
    title="Number of Scientific Articles in 2020"
)
```

## Number of Scientific Publications Over Time China and the US

```python
fig = plt.figure(figsize=(16, 9))

ax = sns.lineplot(
    y = df[df.country_name == "China"].iloc[-1][4:],
    x = list(df.columns[4:].astype(int)),
    color=color_palette[0], linewidth=6, marker="o", markers=True
)


ax = sns.lineplot(
    y = df[df.country_name == "United States"].iloc[-1][4:],
    x = list(df.columns[4:].astype(int)),
    color=color_palette[2], linewidth=6, marker="o", markers=True
)

ax.set(xlabel="", ylabel="")
ax.tick_params("y", labelsize=20, pad=8)
ax.tick_params("x", labelsize=20, pad=16)
ax.yaxis.set_major_formatter(FuncFormatter(lambda x, p: format(int(x), ',')))

b_patch = Patch(color=color_palette[2], label='United States', linewidth=2)
r_patch = Patch(color=color_palette[0], label='China')

leg = plt.legend(
    handles=[b_patch, r_patch],
    bbox_to_anchor=(0.48, 1.09), loc='lower center',
    borderaxespad=0, ncol=3, fontsize=22,
    edgecolor=background_color
)

plt.ylim(0, None)
plt.title("Number of Scientific Publications Over Time", fontsize=32, y=1.22, x=0.48)
plt.tight_layout()

fig.canvas.draw()
data = np.frombuffer(fig.canvas.tostring_rgb(), dtype=np.uint8)
data = data.reshape((fig.canvas.get_width_height()[::-1]) + (3,))
plt.close() 

chart_image = Image.fromarray(data)
chart_image = add_padding_to_chart(chart_image, 30, 20, 30, 30, background_color)
chart_image
```
